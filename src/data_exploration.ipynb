```python
import pandas as pd
import matplotlib.pyplot as plt

# 文件路径
file_path = 'F:/MyCourse(5 delayed 1)/industry/datasets/3633835/EIS data/EIS_state_I_25C01.txt'

# 读取数据
df_eis = pd.read_csv(file_path, sep='\t', engine='python')

# 清理列名
df_eis.columns = df_eis.columns.str.strip()
```

```python
df_eis.head()
```

```python
cycles = df_eis['cycle number'].unique().astype(int)
cycles[0], cycles[-1]
```

```python
# 指定要绘制的周期范围
cycle_start = 1  # 可以根据需要修改这个值
cycle_end = 350   # 可以根据需要修改这个值

# 设置图表
plt.figure(figsize=(8, 6))
plt.xlabel('Real Part of Impedance (Ohm)')
plt.ylabel('Imaginary Part of Impedance (Ohm)')
plt.title('EIS Nyquist Plots for Multiple Cycles')

# 绘制指定范围内的每个周期
for cycle in range(cycle_start, cycle_end + 1):
    cycle_data = df_eis[df_eis['cycle number'] == cycle]
    plt.plot(cycle_data['Re(Z)/Ohm'], cycle_data['-Im(Z)/Ohm'], label=f'Cycle {cycle}')

# plt.legend()
plt.show()
```
