```python
def get_capacity(path):
    df = pd.read_csv(path, sep='\t')
    df_groupby_cycle = df.groupby('cycle number')
    capacity = []
    for cycle, df in df_groupby_cycle:
        capacity.append(df['Capacity/mA.h'].max())
        
    print(f"数据长度：{len(capacity)}")
    plt.figure(figsize=(8, 5))
    sns.scatterplot(capacity, s=10)
    plt.title('Capacity vs Cycle', fontsize=18)
    plt.xlabel('Cycle', fontsize=16)
    plt.ylabel('Capacity/mA.h', fontsize=16)
    plt.show()
    
    np.save(f'{path.split(".")[0]}.npy', capacity)
```
